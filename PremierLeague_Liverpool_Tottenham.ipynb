```python
import pandas as pd
```

```python
df = pd.read_csv(r"results.csv", encoding = "ISO-8859-1")
```

```python
filt = df['HomeTeam'] == 'Liverpool'
filt2 = df['AwayTeam'] == 'Tottenham'
filt3 = df['Season'] > '2000-21'
TotalMatch = df[filt][filt2][filt3]
TotalMatch
```

```python
# Home goal difference
TotalMatch['HGD'] = TotalMatch['FTHG'] - TotalMatch['FTAG']
# Home corner difference
TotalMatch['HCD'] = TotalMatch['HC'] - TotalMatch['AC']
# Home shot difference
TotalMatch['HSD'] = TotalMatch['HS'] - TotalMatch['AS']
# Home shot on target difference
TotalMatch['HSTD'] = TotalMatch['HST'] - TotalMatch['AST']
# Home foul difference
TotalMatch['HFD'] = TotalMatch['HF'] - TotalMatch['AF']
TotalMatch
```

```python
# Win: 0, Draw: 1, Lose: 2

TotalMatch['FTR'] = TotalMatch['FTR'].replace(['H','D','A'], [0, 1, 2])
TotalMatch
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import precision_recall_fscore_support as score, confusion_matrix, roc_auc_score, classification_report, log_loss

from sklearn.neural_network import MLPClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.gaussian_process import GaussianProcessClassifier
from sklearn.gaussian_process.kernels import RBF
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis
from sklearn.linear_model import LogisticRegression
```

```python
features = TotalMatch[['HCD', 'HSD', 'HSTD', 'HFD']]
target = TotalMatch['FTR']
```

```python
X_train,X_test,y_train,y_test = train_test_split(features, target, test_size=0.2, stratify = target)
```

```python
names = ["Nearest Neighbors", "Logistic Regression","Linear SVM", "RBF SVM", "Gaussian Process",
         "Decision Tree", "Random Forest", "Neural Net", "AdaBoost",
         "Naive Bayes", "QDA"]

classifiers = [
    KNeighborsClassifier(3),
    LogisticRegression(),
    SVC(kernel="linear", C=0.025, probability=True),
    SVC(gamma=2, C=1, probability=True),
    GaussianProcessClassifier(1.0 * RBF(1.0)),
    DecisionTreeClassifier(max_depth=5),
    RandomForestClassifier(max_depth=5, n_estimators=10, max_features=1),
    MLPClassifier(alpha=1, max_iter=1000),
    AdaBoostClassifier(),
    GaussianNB(),
    QuadraticDiscriminantAnalysis()]
```

```python
for name, clf in zip(names, classifiers):
        clf.fit(X_train, y_train)
        accuracy = clf.score(X_test, y_test)

        # prediction_proba = clf.predict_proba(X_test)
        # logloss = log_loss(y_test,prediction_proba)
        # precision, recall, fscore, support = score(y_test, prediction)
        # conf_martrix = confusion_matrix(y_test, prediction)
        # clas_report = classification_report(y_test, prediction)

        print(name, accuracy)
```
