```python
import pandas as pd
```

```python
df = pd.read_csv(r"results.csv", encoding = "ISO-8859-1")
```

```python
filt = df['Season'] == '2020-21'
Match_2021 = df[filt]
Match_2021
```

```python
filt = df['Season'] == '2019-20'
Match_1920 = df[filt]
Match_1920
```

```python
filt = df['Season'] == '2018-19'
Match_1819 = df[filt]
Match_1819
```

```python
Match_1819.HomeTeam.unique()
```

```python
Teams_1821 = []

for team in Match_2021.HomeTeam.unique():
    if team not in Teams_1821:
        Teams_1821.append(team)

for team in Match_1920.HomeTeam.unique():
    if team not in Teams_1821:
        Teams_1821.append(team)

for team in Match_1819.HomeTeam.unique():
    if team not in Teams_1821:
        Teams_1821.append(team)
        
print(Teams_1821)
print(len(Teams_1821))
```

```python
Match_2021.AwayTeam.unique()
```

```python
selected_team = "Liverpool"
filt2 = df['HomeTeam'] == selected_team
HomeMatch_2021 = Match_2021[filt2]
HomeMatch_2021

filt3 = HomeMatch_2021['FTHG'] > HomeMatch_2021['FTAG']
HomeWin_2021 = HomeMatch_2021[filt3]
HomeWin_2021

filt4 = HomeMatch_2021['FTHG'] == HomeMatch_2021['FTAG']
HomeDraw_2021 = HomeMatch_2021[filt4]
HomeDraw_2021
```

```python
filt5 = df['AwayTeam'] == selected_team
AwayMatch_2021 = Match_2021[filt5]
AwayMatch_2021 

filt6 = AwayMatch_2021 ['FTHG'] < AwayMatch_2021 ['FTAG']
AwayWin_2021 = AwayMatch_2021 [filt6]
AwayWin_2021

filt7 = AwayMatch_2021 ['FTHG'] == AwayMatch_2021 ['FTAG']
AwayDraw_2021 = AwayMatch_2021 [filt7]
AwayDraw_2021
```

```python
TotalWin_2021 = pd.concat([HomeWin_2021, AwayWin_2021])
TotalWin_2021

TotalDraw_2021 = pd.concat([HomeDraw_2021, AwayDraw_2021])
TotalDraw_2021

winPoint_2021 = len(TotalWin_2021.index) * 3
winPoint_2021

drawPoint_2021 = len(TotalDraw_2021.index)
drawPoint_2021

totalPoint_2021 = winPoint_2021 + drawPoint_2021
totalPoint_2021
```

```python
result = []

for selected_team in Teams_1821:
    count = 0
    filt2 = df['HomeTeam'] == selected_team
    HomeMatch_2021 = Match_2021[filt2]
    #HomeMatch
    filt2 = df['HomeTeam'] == selected_team
    HomeMatch_1920 = Match_1920[filt2]
    #HomeMatch
    filt2 = df['HomeTeam'] == selected_team
    HomeMatch_1819 = Match_1819[filt2]
    #HomeMatch

    filt3 = HomeMatch_2021['FTHG'] > HomeMatch_2021['FTAG']
    HomeWin_2021 = HomeMatch_2021[filt3]
    #HomeWin
    filt3 = HomeMatch_1920['FTHG'] > HomeMatch_1920['FTAG']
    HomeWin_1920 = HomeMatch_1920[filt3]
    #HomeWin
    filt3 = HomeMatch_1819['FTHG'] > HomeMatch_1819['FTAG']
    HomeWin_1819 = HomeMatch_1819[filt3]
    #HomeWin

    filt4 = HomeMatch_2021['FTHG'] == HomeMatch_2021['FTAG']
    HomeDraw_2021 = HomeMatch_2021[filt4]
    #HomeDraw
    filt4 = HomeMatch_1920['FTHG'] == HomeMatch_1920['FTAG']
    HomeDraw_1920 = HomeMatch_1920[filt4]
    #HomeDraw
    filt4 = HomeMatch_1819['FTHG'] == HomeMatch_1819['FTAG']
    HomeDraw_1819 = HomeMatch_1819[filt4]
    #HomeDraw
    
    filt5 = df['AwayTeam'] == selected_team
    AwayMatch_2021 = Match_2021[filt5]
    #AwayMatch
    filt5 = df['AwayTeam'] == selected_team
    AwayMatch_1920 = Match_1920[filt5]
    #AwayMatch
    filt5 = df['AwayTeam'] == selected_team
    AwayMatch_1819 = Match_1819[filt5]
    #AwayMatch

    filt6 = AwayMatch_2021['FTHG'] < AwayMatch_2021['FTAG']
    AwayWin_2021 = AwayMatch_2021[filt6]
    #AwayWin
    filt6 = AwayMatch_1920['FTHG'] < AwayMatch_1920['FTAG']
    AwayWin_1920 = AwayMatch_1920[filt6]
    #AwayWin
    filt6 = AwayMatch_1819['FTHG'] < AwayMatch_1819['FTAG']
    AwayWin_1819 = AwayMatch_1819[filt6]
    #AwayWin

    filt7 = AwayMatch_2021['FTHG'] == AwayMatch_2021['FTAG']
    AwayDraw_2021 = AwayMatch_2021[filt7]
    #AwayDraw
    filt7 = AwayMatch_1920['FTHG'] == AwayMatch_1920['FTAG']
    AwayDraw_1920 = AwayMatch_1920[filt7]
    #AwayDraw
    filt7 = AwayMatch_1819['FTHG'] == AwayMatch_1819['FTAG']
    AwayDraw_1819 = AwayMatch_1819[filt7]
    #AwayDraw
    
    
    TotalWin_2021 = pd.concat([HomeWin_2021, AwayWin_2021])
    #TotalWin
    TotalWin_1920 = pd.concat([HomeWin_1920, AwayWin_1920])
    #TotalWin
    TotalWin_1819 = pd.concat([HomeWin_1819, AwayWin_1819])
    #TotalWin

    TotalDraw_2021 = pd.concat([HomeDraw_2021, AwayDraw_2021])
    #TotalDraw
    TotalDraw_1920 = pd.concat([HomeDraw_1920, AwayDraw_1920])
    #TotalDraw
    TotalDraw_1819 = pd.concat([HomeDraw_1819, AwayDraw_1819])
    #TotalDraw

    winPoint_2021 = len(TotalWin_2021.index) * 3
    #winPoint
    winPoint_1920 = len(TotalWin_1920.index) * 3
    #winPoint
    winPoint_1819 = len(TotalWin_1819.index) * 3
    #winPoint

    drawPoint_2021 = len(TotalDraw_2021.index)
    #drawPoint
    drawPoint_1920 = len(TotalDraw_1920.index)
    #drawPoint
    drawPoint_1819 = len(TotalDraw_1819.index)
    #drawPoint

    totalPoint = winPoint_2021 + winPoint_1920 + winPoint_1819 + drawPoint_2021 + drawPoint_1920 + drawPoint_1819
    
    if selected_team in Match_2021.HomeTeam.unique():
        count += 1
    
    if selected_team in Match_1920.HomeTeam.unique():
        count += 1
        
    if selected_team in Match_1819.HomeTeam.unique():
        count += 1
    
    averagePoint = totalPoint/count
    
    result.append([selected_team, averagePoint])
    #print(selected_team, totalPoint, count, averagePoint)

print(result)
```

```python
# Create the pandas DataFrame with column name is provided explicitly
df_result = pd.DataFrame(result, columns=['Teams', 'Points'])

# print dataframe.
#df_result

df_result.sort_values(by=['Points'], ascending=False, inplace=True)

df_result
```

```python
df_result = df_result.reset_index()

del df_result["index"]
df_result
```

```python
df_result.to_csv("Team_Rank.csv", encoding='ISO-8859-1')
```
