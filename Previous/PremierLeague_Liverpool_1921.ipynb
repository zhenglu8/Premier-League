```python
import pandas as pd
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv(r"results.csv", encoding = "ISO-8859-1")
```

# Home Match

```python
filt_1920 = df['Season'] == '2019-20'
df[filt_1920]
filt2 = df['HomeTeam'] == 'Liverpool'
HomeMatch_1920 = df[filt_1920][filt2]
HomeMatch_1920
```

```python
filt_2021 = df['Season'] == '2020-21'
df[filt_2021]
filt2 = df['HomeTeam'] == 'Liverpool'
HomeMatch_2021 = df[filt_2021][filt2]
HomeMatch_2021
```

```python
TotalHomeMatch = pd.concat([HomeMatch_1920, HomeMatch_2021])
TotalHomeMatch
```

# Convert H (Home Win) to 0 (W)

```python
# Win: 0, Draw: 1, Lose: 2
#df['first_set'] = df['first_set'].replace(['Blue'], 'Green')

TotalHomeMatch['FTR'] = TotalHomeMatch['FTR'].replace(['H','D','A'], [0, 1, 2])
TotalHomeMatch
```

# Home difference

```python
# Home goal difference
TotalHomeMatch['HGD'] = TotalHomeMatch['FTHG'] - TotalHomeMatch['FTAG']
# Home corner difference
TotalHomeMatch['HCD'] = TotalHomeMatch['HC'] - TotalHomeMatch['AC']
# Home shot difference
TotalHomeMatch['HSD'] = TotalHomeMatch['HS'] - TotalHomeMatch['AS']
# Home shot on target difference
TotalHomeMatch['HSTD'] = TotalHomeMatch['HST'] - TotalHomeMatch['AST']
# Home foul difference
TotalHomeMatch['HFD'] = TotalHomeMatch['HF'] - TotalHomeMatch['AF']
TotalHomeMatch
```

# Home goal and home conceded goal

```python
print(TotalHomeMatch['FTHG'].sum())
print(TotalHomeMatch['FTAG'].sum())
```

# Home win

```python
filt3 = TotalHomeMatch['FTHG'] > TotalHomeMatch['FTAG']
TotalHomeWin = TotalHomeMatch[filt3]
TotalHomeWin
#print('Liverpool home win: ' + str(len(HomeWin.index)))
```

```python
filt4 = TotalHomeMatch['FTHG'] == TotalHomeMatch['FTAG']
TotalHomeDraw = TotalHomeMatch[filt4]
TotalHomeDraw
#print('Liverpool home draw: ' + str(len(HomeDraw.index)))
```

# Away Match

```python
filt_1920 = df['Season'] == '2019-20'
df[filt_1920]
filt2 = df['AwayTeam'] == 'Liverpool'
AwayMatch_1920 = df[filt_1920][filt2]
AwayMatch_1920
```

# Convert A (Away Win) to 0 (W)

```python
# Win: 0, Draw: 1, Lose: 2

TotalAwayMatch['FTR'] = TotalAwayMatch['FTR'].replace(['H','D','A'], [2, 1, 0])
TotalAwayMatch
```

```python
features = TotalMatch[['CD', 'SD', 'STD', 'FD']]
```

Target

```python
target = TotalMatch['FTR']
target
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import precision_recall_fscore_support as score, confusion_matrix, roc_auc_score, classification_report, log_loss

from sklearn.neural_network import MLPClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.gaussian_process import GaussianProcessClassifier
from sklearn.gaussian_process.kernels import RBF
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis
from sklearn.linear_model import LogisticRegression
```

```python
X_train,X_test,y_train,y_test = train_test_split(features, target, test_size=0.3, stratify = target)
```

```python
names = ["Nearest Neighbors", "Logistic Regression","Linear SVM", "RBF SVM", "Gaussian Process",
         "Decision Tree", "Random Forest", "Neural Net", "AdaBoost",
         "Naive Bayes", "QDA"]

classifiers = [
    KNeighborsClassifier(3),
    LogisticRegression(),
    SVC(kernel="linear", C=0.025, probability=True),
    SVC(gamma=2, C=1, probability=True),
    GaussianProcessClassifier(1.0 * RBF(1.0)),
    DecisionTreeClassifier(max_depth=5),
    RandomForestClassifier(max_depth=5, n_estimators=10, max_features=1),
    MLPClassifier(alpha=1, max_iter=1000),
    AdaBoostClassifier(),
    GaussianNB(),
    QuadraticDiscriminantAnalysis()]
```

```python
for name, clf in zip(names, classifiers):
        clf.fit(X_train, y_train)
        accuracy = clf.score(X_test, y_test)

        # prediction_proba = clf.predict_proba(X_test)
        # logloss = log_loss(y_test,prediction_proba)
        # precision, recall, fscore, support = score(y_test, prediction)
        # conf_martrix = confusion_matrix(y_test, prediction)
        # clas_report = classification_report(y_test, prediction)

        print(name, accuracy)
```
